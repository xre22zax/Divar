# Question 2 

```python
# Count number of items in each cell of 'tokens' column
load['token_length'] = load['tokens'].apply(len)/10

load.head()
```

```python
# Filter unique source_event_id where token_length < 10
filtered_ids = load[load['token_length'] < 10]['source_event_id'].unique()

# Calculate percentage
percentage_dark_query = len(filtered_ids) / len(load['source_event_id'].unique()) * 100

print(f"Percentage dark query is: {percentage_dark_query:.2f}%")
```

```python
# Set a counter to find out number of load post that've been clicked
counter = 0

# Itrate through tokens check if post token exist
for i in click['post_token']:
    found = False
    for t in load['tokens']:
        if i in t:
            found = True
            break
    if not found:
        counter += 1
        
print(counter)
```

```python
# Calculate Click bounce rate
bounce_rate = (counter/len(load)*100)
print(f"Click bounce rate {bounce_rate:.2f}%")
```

# Question 3

```python
# Calculate len of all tokens
total_indexes = 0

for tokens in load['tokens']:
    total_indexes += len(tokens)

    
# Calculate What is percentage of clicks in comparison to all clicks
click_load_percentage = (len(click)/total_indexes)*100

print(f"Percentage of Clicks vs Load ADs is: {click_load_percentage:.2f}%")
```

```python
# Dictionary to store the first click index for each device_id
First_click_index = {}

# Filter the load_df to include only device_ids that are present in click_df
filtered_load_df = load[load['device_id'].isin(click['device_id'])]

# Iterate through each row in filtered load df
for idx, load_row in filtered_load_df.iterrows():
    device_id = load_row['device_id']
    tokens = load_row['tokens']
    
    # Find the post_token for this device_id in click_df
    post_token = click[click['device_id'] == device_id]['post_token'].values[0]
    
    try:
        # Find the index of post_token in tokens list
        token_index = tokens.index(post_token)
        # Store the result in the dictionary
        First_click_index[device_id] = token_index
    except ValueError:
        # If post_token is not in the tokens list
        First_click_index[device_id] = -1

# Filter out entries with -1 values
First_click_index = {k: v for k, v in First_click_index.items() if v > -1}

print('Index of first post_token for each device_id with valid indices:')
print(First_click_index)
```

```python
First_click_index = {}

# Filter the load to include only device_ids that are present in click
filtered_load = load[load['device_id'].isin(click['device_id'])]

# Iterate through each row in filtered load
for idx, load_row in filtered_load.iterrows():
    device_id = load_row['device_id']
    tokens = load_row['tokens']
    
    # Find the post_token for this device_id in click
    post_token = click[click['device_id'] == device_id]['post_token'].values[0]
    
    try:
        # Find the index of post_token in tokens list
        token_index = tokens.index(post_token)
        # Store the result in the dictionary
        if device_id in First_click_index:
            First_click_index[device_id].append(token_index)
        else:
            First_click_index[device_id] = [token_index]
    except ValueError:
        # If post_token is not in the tokens list
        if device_id not in First_click_index:
            First_click_index[device_id] = [-1]

# Filter out entries with -1 values
First_click_index = {k: [v for v in vals if v > -1] for k, vals in First_click_index.items()}

# Create a DataFrame from the dictionary
first_click_df = pd.DataFrame(list(First_click_index.items()), columns=['device_id', 'indices'])

# Explode the DataFrame so each index is in its own row
exploded_df = first_click_df.explode('indices')

# Convert the indices to numeric
exploded_df['indices'] = pd.to_numeric(exploded_df['indices'])

# Group by device_id and calculate the mean
result = exploded_df.groupby('device_id')['indices'].mean()

# Convert the result back to a dictionary and filter out NaN values
average_indices = result.dropna().to_dict()

print('Average index of first post_token for each device_id:')
print(average_indices)
```

```python
import matplotlib.pyplot as plt

First_click_index = {}

# Filter the load to include only device_ids that are present in click
filtered_load = load[load['device_id'].isin(click['device_id'])]

# Iterate through each row in filtered load
for idx, load_row in filtered_load.iterrows():
    device_id = load_row['device_id']
    tokens = load_row['tokens']
    
    # Find the post_token for this device_id in click
    post_token = click[click['device_id'] == device_id]['post_token'].values[0]
    
    try:
        # Find the index of post_token in tokens list
        token_index = tokens.index(post_token)
        # Store the result in the dictionary
        if device_id in First_click_index:
            First_click_index[device_id].append(token_index)
        else:
            First_click_index[device_id] = [token_index]
    except ValueError:
        # If post_token is not in the tokens list
        if device_id not in First_click_index:
            First_click_index[device_id] = [-1]

# Filter out entries with -1 values
First_click_index = {k: [v for v in vals if v > -1] for k, vals in First_click_index.items()}

# Create a DataFrame from the dictionary
first_click_df = pd.DataFrame(list(First_click_index.items()), columns=['device_id', 'indices'])

# Explode the DataFrame so each index is in its own row
exploded_df = first_click_df.explode('indices')

# Convert the indices to numeric
exploded_df['indices'] = pd.to_numeric(exploded_df['indices'])

# Group by device_id and calculate the mean
result = exploded_df.groupby('device_id')['indices'].mean()

# Convert the result back to a dictionary and filter out NaN values
average_indices = result.dropna().to_dict()

# Calculate the percentage of device_ids with average click rate less than 3
below_3_count = sum(1 for avg in average_indices.values() if avg < 3)
total_count = len(average_indices)
percentage_below_3 = (below_3_count / total_count) * 100
percentage_above_3 = 100 - percentage_below_3

# Pie chart
labels = ['Below 3', '3 and Above']
sizes = [percentage_below_3, percentage_above_3]

plt.figure(figsize=(8, 8))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#ff9999','#66b3ff'])
plt.title('Percentage of Device IDs by Average Click Rate')
plt.legend(labels, loc="best")
plt.axis('equal')
plt.show()
```
